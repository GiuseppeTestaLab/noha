```python
import pandas as pd
import numpy as np
import glob
```

```python
output_path = '/group/testa/Project/EndPoints/TPSSU/analysis/cytoplasmatic_measures/'
```

```python
files = glob.glob(output_path + '*')
files
```

```python
import pickle

def load_pickle(filepath):
    with open(filepath, "rb") as f:
        return pickle.load(f)
    
cyto_dict = {}
for file in files:
    if file.endswith('.pkl'):
        cyto_mask_measures = load_pickle(file)
        cyto_mask_measures_df = pd.DataFrame.from_dict(cyto_mask_measures).T
        cyto_mask_measures_df['area'] = cyto_mask_measures_df['area'].apply(lambda x: x[0] if x is not None else np.nan)
        cyto_mask_measures_df['label'] = cyto_mask_measures_df['label'].apply(lambda x: x[0] if x is not None else np.nan)
        cyto_mask_measures_df['intensity_mean'] = cyto_mask_measures_df['intensity_mean'].apply(lambda x: x[0] if x is not None else np.nan)
        cyto_mask_measures_df['intensity_min'] = cyto_mask_measures_df['intensity_min'].apply(lambda x: x[0] if x is not None else np.nan)
        cyto_mask_measures_df['intensity_max'] = cyto_mask_measures_df['intensity_max'].apply(lambda x: x[0] if x is not None else np.nan)
        cyto_dict[file] = cyto_mask_measures_df
```

```python
cyto_df = pd.concat(cyto_dict).reset_index()
```
